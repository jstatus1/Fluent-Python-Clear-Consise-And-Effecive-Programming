## ABC or Abstract Base Class
- part of the abc module and is used to define classes that are meant to be inherited from, but not instantiated directly
- use ABC because it helps you define a common interface for a group of related classes. Enforce that certain methods must be implemented in subclasses

```python
from abc import ABC, abstractmethod

# You do not want other devs to instantiate vehicle class
class Vehicle(ABC):
    @abstractmethod
    def go(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):
    def go(self):
        print("You Drive The Car")

    def stop(self):
        print("You Stop The Car")

car = Car()
car.go()
car.stop()
```

- collections.abc module provides the Mapping and MutableMapping ABCs

What is hashable?

- an object is hashable if it has a hash code which never changes during its lifetime
- it can be compared to other objects. 
- frozenset is always hashable bc every element it contains must be hashable
- tuple is hashable only if all its items are hashable
- most commonly used mapping types in Python: dict, defaultdict, and OrderedDict

```python
tt = (1, 2, (30, 40))
hash(tt)
```
