# Chapter 1: The Python Data Model

- Collections.namedtuple isused to construct a simple class to represent individual cards
- use namedtuple to build classes ofobjects that are jut bundles of attributes with no custom methods (ex db record)

```python
import collections

Card = collections.namedtuple('Card', ['rank', 'suit'])

class FrenchDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self):
        self._cards = [Card(rank, suit) for suit in self.suits  
                                        for rank in self.ranks]
    
    def __len__(self):
        return len(self._cards)

    def __getitem__(self, position):
        return self._cards[position]
```

```python
deck = FrenchDeck()
len(deck)


# Get Item From Deck
deck[-1]
```
