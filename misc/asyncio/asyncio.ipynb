## Event Loop

- everything in asyncio happens relative to the event loop
- you can think about it like it is the orchestrator that manages resources
- event loop contains a queue of jobs (tasks, work orders, callbacks, etc). Jobs are added directly by you, and some indirectly by asyncio
- the event loop pops a job from the queue, and executes. After execution it is returned back to the event loop
- Effective overall execution relies on tasks sharing well. Greedy tasks could hog control and leave other tasks to starve 

```python
import asyncio

# This creates an event-loop and indefinetly cycles through its queue of tasks.
event_loop = asyncio.new_event_loop()
#event_loop.run_forever()

print(event_loop)
```

- example of an asynchronous f(x) or coroutine function. Has keyword 'async'

```python
async def special_fella(magic_number: int):
    print(
        "I am a super special function. Far cooler than that printer. By the way, "
        f"my lucky number is: {magic_number}."
    )
```

- calling async functions return coroutine object. Does not execute the function
- coroutine object (aka coroutines)
    - represents the function's body or logic
    - has to be explicitly started
    - coroutines can be paused & resumed at various points within the f(x)'s body. This ability to pause and resume allows for async behaviors
    - were built on top of the scaffolding of generators and generator f(x). Generator f(x) is a function that yields


```python
special_fella(magic_number=3)
```

# What is a Generator?
- lets you iterate over values lazily (one at a time) instead of loading everything into memory at once
- list -> gives you everything upfront
- generator -> gives you values on demand

# What is a Generator Function?
- just a function that uses the keyword yield instead of return
- it does not run immediately
- returns a generator object
- execution pauses and resumes each time you iterate

```python
def count_up_to(n):
    for i in range(n):
        yield i

gen = count_up_to(3)
print(next(gen))
```

```python
print(next(gen))
```

```python
print(next(gen))
```
