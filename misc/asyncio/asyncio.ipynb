## Event Loop

- everything in asyncio happens relative to the event loop
- you can think about it like it is the orchestrator that manages resources
- event loop contains a queue of jobs (tasks, work orders, callbacks, etc). Jobs are added directly by you, and some indirectly by asyncio
- the event loop pops a job from the queue, and executes. After execution it is returned back to the event loop
- Effective overall execution relies on tasks sharing well. Greedy tasks could hog control and leave other tasks to starve
