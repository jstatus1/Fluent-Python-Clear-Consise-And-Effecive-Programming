## Listcomps vs Map and Filter

- map() takes a function and collection, and applies that function to each element
- list() is a constructor that converts any iterable into a list. Commonly used to materialize iterators like map() or generators into concrete data structures
- ord() is built in Python f(x) that returns the unicode code points (integer) of a character
    - ord(): char -> int
    - chr(): int -> char
- filter() applies a function to each item and returns only those where the function is True

- listcomp or list comprehension is a compact way to create lists in python

Basic Syntax:

[expression for item in iterable]

- list comprehension is optimized by Python's bytecode and avoids calling python f(x) for every element. map(lamba...) has to call the lambda repeatedly

- Bytecode is the low-level, intermediate instructions that Python generates from your code before executing it.

How Python runs your code:
your python code (.py) -> compiled into bytecode -> executed by Python Virtual Machine (PVM)

Why does Bytecode matter?

- Performance: python doesn't recompile everything everytime
- Portability: bytecode runs on any system with Python
- List comprehension are faster bc there are less bytecode instructions. NO REPEATING F(x)

# Generator Expressions

- General Expressions or genexp saves memory bc it yields items one by one using the iterator protocol instead of building a whole lsit just o feed another constructor
- GenExps use the same syntax as listcomps, but are enclosed in parenteses instead of brackets

Example 2-6 uses a genexp with a Cartesian Product to print out a rost of T-Shirts of 2 colors in the 3 sizes. The 6 item list of T-Shirt is never built in memory: the generator expression feeeds the for loop producing one item at a time. If 2 list in the Cartesian Product had thousands of items, using a generator expression would save time in that it would not have to build a list with millions of items just ot feed the for loop.

```python
colors = ['black', 'white']
sizes = ['S','M','L']

for tshirt in (f'{c} {s}' for c in colors for s in sizes):
    print(tshirt)
```

```python
# Huge memory usage
pairs = [(x, y) for x in range(10000) for y in range(10000)]

# Tiny memory usage
pairs = ((x, y) for x in range(10000) for y in range(10000))
```

### Nested Unpacking

```python
metro_areas = [
    ('Tokyo', 'JP', 36.933, (35.689722, -139.691667)),
    ('Delhi NCR', 'IN', 21.933, (28.689722, -77.691667)),
    ('Mexico City', 'MX', 20.933, (19.689722, -99.691667))
]

def main():
    print(f'{"":15} | {"latitude":>9} | {"longitude":>9}')

    for name, _, _, (lat, lon) in metro_areas:
        if lon <= 0:
            print(f'{name:15} | {lat:9.4f} | {lon:9.4f}')

if __name__ == '__main__':
    main()
```

## Pattern Matching With Sequences

- comes from structural pattern matching feature from python 3.1 and lets you match the shape and contents of a sequence in a clean readable way using the match/case syntax
- this is great if you are working with variable length lists

```python
def check_sequence(seq):
    match seq:
        case [1, 2, 3]:
            return "Exact match"
        case [1, x, 3]:
            return f"Middle value is {x}"
        case [1, *rest]:
            return f"Starts with 1, rest is {rest}"
        case _:
            return "No match"
```

```python
def handle_command(self, message):
    match message:
        case ['BEEPER', frequency, times]:
            self.beep(times, frequency)
        case ['NECK', angle]:
            self.rotate()
```

Mutable vs Immutable:
- mutable and immutable describe whether an object can be changed after it's created
- immutable can not change. An immutable object can not be modified in place. If you "change" it, Python creates a new object

Common Immutable Types:
- int
- float
- str
- tuple
- bool

```python
x = 10
print(id(x)) # memory address


x = x+5         # create a NEW object
print(id(x))    # different address
```

- mutable object can be modified in place without creating a new object
- common mutable types:
    - list
    - dict
    - set
    - most custom objects

```python
lst = [1, 2, 3]

print(id(lst))

lst.append(4)   # modifies SAME object

print(id(lst))  # same address
```

# Augmeted Assignment With Sequences


- += or __iadd__ (in place addition), if not implemented will fall back to __add__
- in case of mutable sequences (list, bytearray, array.array), it will be changed in place and is similar to a.extend(b)
- when 'a' does not implement __iadd__, the expression a += b

```python
l = [1,2,3]
id(l)

l += [1]
print(l)
print(f"id: {id(l)} | {l}")

l *= 2
print(l)
print(f"id: {id(l)} | {l}")
```

# list.sort vs the sorted built-in

- list.sort sorts a list in place, without amking a copy
- functions or methods that change an object in place should return None to make it clear to the caller that the receiver was changed and no new object was created

- sorted built in function creates a new list and returns it
- accepts any iterable object as an argument (includes immutable sequences and generators)


- Both list.sort and sorted take two optional, keyword only arguments:
    - reverse: if True, the items are returned in descending order (i.e. by reversing the comparison of the items.)
    - key: one argument function that will be applied to each item to produce its soritng key. When sorting a list of strings, key=str.lower can be used to perform a case insensitive sort 

```python
fruits = ['grape', 'raspberry', 'apple', 'banana']

print(sorted(fruits))
print(fruits)

print(sorted(fruits, reverse=True))
print(sorted(fruits, key=len))
print(sorted(fruits, key=len, reverse=True))
print(fruits)

print(fruits.sort())
print(fruits)
```

Ex 2-19 Create, Saving, and Loading A Large Array of Floats

```python
from array import array #import the array type
from random import random

# Create an array of double-precision floats (typecode 'd') from any iterable objects in this case, a generator expression
floats = array('d', (random() for i in range(10**7)))

# Inspect the last element of the array
floats[-1]
```

```python
# Save the array to a binary file
fp = open('floats.bin', 'wb')
floats.tofile(fp)
```

```python
fp.close()
```

- array.tofile and array.fromfile are easy to use. It takes .1 seconds for array.fromfile to load 10 mill double-precision floats from binary file. 60x faster than reading the number from a text file bc txt files involves parsing each line with float built in. 
- array.tofile is 7x faster than writing one float per line in a text file. 
- this is faster because they read/write raw bytes directly instead of converting numbers to and from text

```python
# Create an empty array of doubles
floats2 = array('d')

# Read 10M numbers from the binary file
fp = open('floats.bin', 'rb')
floats2.fromfile(fp, 10**7)
fp.close()

# Inspect the last number in the array
floats2[-1]

#Verify that the contents of the arrays match
floats2 == floats
```

```python
print(id(floats))
print(id(floats2))
```

# Deques and Queues

- .append and .pop methods make list usable as a stack or a queue
- using .pop at the 0 index is costly bc the entire list must be shifted in memory
- collecitons.deque is a thread safe double ended queue designed for fast inserting and removing from both ends
- removing items from the middle of a deque is not as fast. Optimized for appending and popping from the ends
- append and popleft operations are atomic, deque is safe to use as a FIFO queue in multithreaded apps without the need for locks

- maxlen argument sets the max number of items allowed in this instance of deque; this sets a read only maxlen instance attribute

```python
from collections import deque

dq = deque(range(10), maxlen=10)
dq
```
