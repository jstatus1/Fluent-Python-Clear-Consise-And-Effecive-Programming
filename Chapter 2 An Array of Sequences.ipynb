# Chapter 2: An Array Of Sequences

Topics Covered:
- List comprehensions and the absic of generator expressions
- Using tuples as records vs using tuples as immutable lists
- Sequence from slices and writing to slices
- Specialized sequence types, arrays and queues

Built In Sequences:
- Container Sequence: 
    - can hold items of different types, includes nested containers. Ex: list, tuple, and collections.deque

- Flat Sequences:
    - hold items of one simple type. Ex: str, bytes, and array.array
