## Lambda

- Lambdas in python is small, anonymous (unnamed) f(x) defined in a single line

```python
add = lambda a, b: a + b
print(add(2,3))
```

- Common use cases for lambdas:

1. with functions like map, filter, sorted

```python
nums = [1, 2, 3, 4]

squared = list(map(lambda x: x**2, nums))

print(squared)
```

2. Sorting With Custom Logic

```python
pairs = [(1,3), (2,1), (4,2)]

pairs.sort(key=lambda x: x[1])

print(pairs)
```

## Listcomps vs Map and Filter

- map() takes a function and collection, and applies that function to each element
- list() is a constructor that converts any iterable into a list. Commonly used to materialize iterators like map() or generators into concrete data structures
- ord() is built in Python f(x) that returns the unicode code points (integer) of a character
    - ord(): char -> int
    - chr(): int -> char
- filter() applies a function to each item and returns only those where the function is True

- listcomp or list comprehension is a compact way to create lists in python

Basic Syntax:

[expression for item in iterable]

- list comprehension is optimized by Python's bytecode and avoids calling python f(x) for every element. map(lamba...) has to call the lambda repeatedly

- Bytecode is the low-level, intermediate instructions that Python generates from your code before executing it.

How Python runs your code:
your python code (.py) -> compiled into bytecode -> executed by Python Virtual Machine (PVM)

Why does Bytecode matter?

- Performance: python doesn't recompile everything everytime
- Portability: bytecode runs on any system with Python
- List comprehension are faster bc there are less bytecode instructions. NO REPEATING F(x)

# Generator Expressions

- General Expressions or genexp saves memory bc it yields items one by one using the iterator protocol instead of building a whole lsit just o feed another constructor
- GenExps use the same syntax as listcomps, but are enclosed in parenteses instead of brackets

Example 2-6 uses a genexp with a Cartesian Product to print out a rost of T-Shirts of 2 colors in the 3 sizes. The 6 item list of T-Shirt is never built in memory: the generator expression feeeds the for loop producing one item at a time. If 2 list in the Cartesian Product had thousands of items, using a generator expression would save time in that it would not have to build a list with millions of items just ot feed the for loop.

```python
colors = ['black', 'white']
sizes = ['S','M','L']

for tshirt in (f'{c} {s}' for c in colors for s in sizes):
    print(tshirt)
```

```python
# Huge memory usage
pairs = [(x, y) for x in range(10000) for y in range(10000)]

# Tiny memory usage
pairs = ((x, y) for x in range(10000) for y in range(10000))
```
